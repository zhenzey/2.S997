```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FormatStrFormatter
import matplotlib as mpl
from matplotlib import font_manager as fm
import os
import re
```

```python

print(mpl.rcParams['datapath'])
fname = os.path.join(mpl.rcParams['datapath'], 'fonts', 'ttf', 'helvetica.ttf')
prop = fm.FontProperties(size=13, fname=fname)

# Fix initialization time and generation number, change populaiton size
num_pop_size = 4
scores = []
pop_size = []
for i in range(num_pop_size):
    pop_size.append(int(4 * np.power(5, i)))
    f = open("./scores/pop_scores/pop_" + str(int(4 * np.power(5, i))) + "/scores.txt")
    lines = f.readlines()
    score = float(re.findall(r"[-+]?\d*\.\d+|\d+", lines[-1])[0])
    scores.append(score)

plt.subplots(figsize = (6, 6))
plt.plot(pop_size, scores, marker='o')
plt.xlabel("population size", font_properties=prop)
plt.ylabel("overall score", font_properties=prop)
plt.xticks(font_properties=prop)
plt.yticks(font_properties=prop)
plt.title("gen_num=1000, init_time=16", font_properties=prop)
plt.savefig("pop.png", dpi=300)
plt.show()




# Fix initialization time and generation number, change populaiton size
num_gen_size = 4
scores = []
gen_num = []
for i in range(num_gen_size):
    gen_num.append(int(8 * np.power(5, i)))
    f = open("./scores/gen_scores/gen_" + str(int(8 * np.power(5, i))) + "/scores.txt")
    lines = f.readlines()
    score = float(re.findall(r"[-+]?\d*\.\d+|\d+", lines[-1])[0])
    scores.append(score)

plt.subplots(figsize = (6, 6))
plt.plot(gen_num, scores, marker='o')
plt.xlabel("generation number", font_properties=prop)
plt.ylabel("overall score", font_properties=prop)
plt.xticks(font_properties=prop)
plt.yticks(font_properties=prop)
plt.title("pop_size=500, init_time=16", font_properties=prop)
plt.savefig("gen.png", dpi=300)
plt.show()



num_init_time = 3
scores = []
init_time = []
for i in range(num_init_time):
    init_time.append(int(1 * np.power(4, i)))
    f = open("./scores/init_scores/init_" + str(int(1 * np.power(4, i))) + "/scores.txt")
    lines = f.readlines()
    score = float(re.findall(r"[-+]?\d*\.\d+|\d+", lines[-1])[0])
    scores.append(score)

plt.subplots(figsize = (6, 6))
plt.plot(init_time, scores, marker='o')
plt.xlabel("initialization time", font_properties=prop)
plt.ylabel("overall score", font_properties=prop)
plt.xticks(font_properties=prop)
plt.yticks(font_properties=prop)
plt.title("pop_size=500, gen_num=1000", font_properties=prop)
plt.savefig("init.png", dpi=300)
plt.show()

```
